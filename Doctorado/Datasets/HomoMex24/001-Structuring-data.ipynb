```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
data_path = 'Task3/Raw/etiquetado_2/'
```

```python
# Cargar los datos
songs = pd.read_csv(f'{data_path}sampled_songs_part_2.csv')
labels = pd.read_csv(f'{data_path}labels.csv')

# Crear una nueva columna en songs que contenga los índices de las filas
songs['merge_key'] = songs.index

# Cambiar el tipo de datos de 'song_id' en labels a int si es necesario
labels['song_id'] = labels['song_id'].astype(int)

# Unir los datos usando la nueva columna 'merge_key' y 'song_id' de labels
merged_data = pd.merge(songs, labels, left_on='merge_key', right_on='song_id')

# Seleccionar solo las columnas necesarias: song_id de songs, lyrics y label
final_data = merged_data[['song_id_x', 'lyrics', 'label']]  # Asumiendo que 'lyrics' está en songs
final_data.rename(columns={'song_id_x': 'song_id'}, inplace=True)  # Renombrar para claridad

# Guardar el DataFrame resultante en un nuevo archivo CSV
final_data.to_csv('final_songs_data.csv', index=False)
```
